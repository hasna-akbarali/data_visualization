```python
import seaborn as sns
```

## Built-in DataSet called tips in seaborn

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

### Considering tip as dependent feature and the rest as independent 

```python
sns.heatmap(df.corr(),annot=True)
```

### Univariate Analysis - sns.jointplot()

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### When more that 2 features need to be analyzed we use sns.pairplot()

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.distplot(df['tip'])
```
